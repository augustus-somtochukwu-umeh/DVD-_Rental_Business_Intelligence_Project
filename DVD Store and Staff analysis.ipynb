```python
df3 = pd.read_csv('Customer Behaviour and Rentention Dataset')
df3.head()
```

```python
repeat_customers = df3[df3['total_no_of_rentals'] > 1]

repeat_percentage = (len(repeat_customers) / len(df3)) * 100

print(f"{repeat_percentage:.2f}% of customers are repeat renters")
```

```python
average_spending = df3['total_amount_spent'].mean()
print(f"Average spending per customer: ${average_spending:.2f}")
```

```python
df3['first_rental_date'] = pd.to_datetime(df3['first_rental_date'])
df3['most_recent_rental_date'] = pd.to_datetime(df3['most_recent_rental_date'])

df3['avg_gap_days'] = (df3['most_recent_rental_date'] - df3['first_rental_date']).dt.days / (df3['total_no_of_rentals'] - 1)
df3['avg_gap_days'] = df3['avg_gap_days'].fillna(0)

overall_avg_gap = df3['avg_gap_days'].mean()

print(f"Average time gap between rentals: {overall_avg_gap:.2f} days")
```

```python
churn_threshold_days = 180  

reference_date = df3['most_recent_rental_date'].max()

df3['is_churn_risk'] = (
    (reference_date - df3['most_recent_rental_date']).dt.days > churn_threshold_days
)

num_churn_risk = df3['is_churn_risk'].sum()
total_customers = len(df3)

churn_rate = num_churn_risk / total_customers
df3.head()
```
