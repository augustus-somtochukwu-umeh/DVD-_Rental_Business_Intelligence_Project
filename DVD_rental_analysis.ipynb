```python
df4.rename(columns={'title': 'film_title', 'name': 'category', 'total_rentals': 'total_num_of_rental'}, inplace=True)
df4.head()
```

```python
merged_df1 = pd.merge(df1, df4, on=['category', 'film_title', 'last_rental_date', 'total_num_of_rental'], how='inner')
merged_df1.head()
```

```python
merged_df1.info()
```

```python
merged_df1['last_rental_date'] = pd.to_datetime(merged_df1['last_rental_date'], errors='coerce', dayfirst=False)
merged_df1.info()
```

```python
rental_distribution = px.histogram(
    merged_df1,
    x='total_rental_revenue',
    nbins=30,
    title='Rental Revenue Distribution'
)
rental_distribution.show()
```

```python
fig_rental_duration_distribution = px.histogram(
    merged_df1,
    x='avg_rental_duration',
    nbins=10,
    title='Rental Duration Distribution'
)
fig_rental_duration_distribution.show()
```

```python
#outliers detection
num_cols = ['avg_rental_duration', 'total_rental_revenue', 'total_num_of_rental']
for col in num_cols:
    Q1 = merged_df1[col].quantile(0.25)
    Q3=merged_df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1-1.5*IQR
    upper_bound = Q3+1.5*IQR
    merged_df1 = merged_df1[(merged_df1[col] >= lower_bound) & (merged_df1[col] <= upper_bound)]
```

```python
merged_df1['year_month'] = merged_df1['last_rental_date'].dt.to_period('M').astype(str)
merged_df1.head()
```

```python
trend = merged_df1.groupby('year_month').agg(
    total_rentals=('total_num_of_rental', 'sum'),
    total_revenue=('total_rental_revenue', 'sum')
).reset_index()

fig_rental_and_revenue_distribution = px.line(
    trend,
    x='year_month',
    y=['total_rentals','total_revenue'],
    title='Rental and Revenue Trend Over Time'
)

fig_rental_and_revenue_distribution.show()
```

# Business Recommendations and Strategies

# Revenue Growth Potential

1.  The 3 film categories the company should invest more in are the sports, comedy, and games. This is because these three categories are the film categories that generate the highest revenue. On the otherhand, some film categories such as music, classics and drama needs to be discontinued as a result of generating lowest rental revenue.

2. Some films generate high revenue but moderate rental numbers → high willingness to pay per rental. Other films have many rentals but low revenue → potentially low pricing or short rental duration.
Recommendation:
Dynamic pricing by popularity and revenue: Charge slightly higher rental fees for high-demand, high-revenue films. Offer discounts on low-demand films to stimulate rentals.
Bundle or package rentals
Promotional pricing by seasonality

# Customer Retention Strategy
3. Actions to retain high-value customers:
Innovation and learning
Offer personalized offers, e.g., “Your favorite category is discounted this week.”
Give loyalty points for frequent rentals to encourage continued engagement.

4. Re-engaging inactive or churn-risk customers:
Send reminder emails or notifications about new releases or popular films.
Offer special discounts for returning after inactivity.
Highlight “missed films” or “You haven’t watched these” campaigns.

5. Loyalty program design:
Tiered rewards:
Silver: 1–5 rentals/month → small discounts
Gold: 6–10 rentals/month → bigger discounts + early access to new films
Platinum: 10+ rentals/month → free rentals, exclusive offers
Include bonus points for referrals to attract new customers.

# Store & Staff Performance Improvement
1) Company-wide store practices:
Replicate practices from highest revenue stores (e.g., display layouts, promotions, staff-customer interactions).
Track performance metrics like revenue per rental and total rentals per staff.

2) Training or staffing changes:
Train staff in customer engagement and upselling (encouraging customers to rent more or premium films).
Assign staff based on efficiency metrics: top-performing staff should handle peak hours.

3) Staffing schedule optimization:
Schedule more staff during peak rental periods (weekends, evenings, paydays).
Reduce staffing in low-traffic hours to save costs.

# Market Expansion and Localization
1. India, China, United States, and Japan show the highest growth revenue potential as they have the highest total revenue
2. Stock local favorites and top-performing categories per city.
   Offer genre-specific promotions (family films, action, documentaries) based on local rental trends.
3. Localized marketing strategies:
    Use regional campaigns highlighting most rented films in that city.
    Partner with local events or cinema promotions to attract new customers.
    Implement geo-targeted digital campaigns for inactive customers in specific regions.

# Inventory Optimization
1. Texas watch and Oklahoma Jumanji are the movies with the lowest total revenue and should be removed from the inventory due to low performance. This helps free storage and reduce costs.
2. Inventory by store/region:
Stock popular films locally based on regional rental patterns.
Stores in high-demand cities should carry more copies of top-performing films.
Use historical data to forecast seasonal demand and adjust stock dynamically.
